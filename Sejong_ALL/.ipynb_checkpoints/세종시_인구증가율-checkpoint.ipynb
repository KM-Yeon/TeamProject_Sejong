```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rc('font', family='NanumBarunGothic')
```

```python
df = pd.read_csv('/content/drive/MyDrive/세종시/data/인구증가율세종.csv')
df = df.T
df
```

```python
df.info()
```

```python
fig = plt.figure(figsize=(15,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
 
ax.plot(df.index,df[0], marker='o', color='r') ## 선그래프 생성
plt.ylim(0,100)
plt.xlabel('년도')
plt.ylabel('증가율(%)')
plt.xticks(rotation=45) ## x축 눈금 라벨 설정 - 40도 회전 
plt.title('세종시 년도 별 인구 증가율',fontsize=20) ## 타이틀 설정
plt.show()
```
