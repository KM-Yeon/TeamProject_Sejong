```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic')
```

```python
df = pd.read_csv('/content/drive/MyDrive/세종시/data/6.세종시_오피스텔(매매)_실거래가.csv')
df.head()
```

```python
df['동네'] = df.시군구.str.split(' ').str[1]
df.head()
```

```python
# 불필요한 컬럼 삭제

df = df.drop(['계약년월', '계약일','시군구','본번','부번','단지명','해제사유발생일'],axis=1)
df
```

```python
# object형인 거래금액 int 형으로 바꾸기 

df['거래금액(만원)'] = df['거래금액(만원)'].str.replace(',', '').astype('int64')
df.head()
```

```python
df['동네'].unique()
```

```python
df_oh = pd.get_dummies(data=df, columns=['동네'])
df_oh
```

```python
#컬럼 순서 바꾸기

df_oh.columns
```

```python
df_oh = df_oh[['전용면적(㎡)', '층', '건축년도', '동네_나성동', '동네_도담동', '동네_부강면',
       '동네_어진동', '동네_조치원읍', '거래금액(만원)']]
df_oh.head()
```

```python
plt.figure(figsize=(16,14))
sns.heatmap(df_oh.corr(), annot=True, cmap = 'RdYlBu_r',vmin = -1, vmax = 1)
plt.show()
```

- 원룸구조의 오피스텔이므로 층수 요소는 거래금액에 크게 영향을 끼치지 않는 것으로 보임.
- '나성동' 은 왜 영향을 많이 끼치는지 모르겠음 ㅇㅅㅇ
