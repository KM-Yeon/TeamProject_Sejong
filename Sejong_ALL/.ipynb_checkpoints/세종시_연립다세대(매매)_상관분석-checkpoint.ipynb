```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic')
```

# 파일 불러오기

```python
df = pd.read_csv('/content/drive/MyDrive/세종시/data/4.세종시_연립다세대(매매)_실거래가.csv')
df.head()
```

# object -> int 형으로 변환

```python
df['거래금액(만원)'] = df['거래금액(만원)'].str.replace(',', '').astype('int64')
df.head()
```

# '동네' 컬럼 추가

```python
df['동네'] = df.시군구.str.split(' ').str[1]
df
```

# 불필요한 컬럼 삭제
- 가격에 영향을 주지 않는 컬럼 삭제

```python
df2 = df.drop(['계약년월', '계약일','시군구','본번','부번','건물명','해제사유발생일'],axis=1)
df2
```

#One hot Encoding

```python
df_oh = pd.get_dummies(df2['동네'])
df_oh.head()
```

```python
df3 = pd.concat([df2,df_oh],axis=1)
df3
```

# 상관관계 heatmap

```python
df3=df3[['전용면적(㎡)', '대지권면적(㎡)', '층', '건축년도', '동네', '금남면', '도담동', '반곡동', '보람동', '부강면', '소담동', '소정면', '연기면', '연서면', '장군면', '전동면', '전의면', '조치원읍' , '거래금액(만원)']]
```

```python
plt.figure(figsize=(20,18))
sns.heatmap(df3.corr(), annot=True, cmap = 'RdYlBu_r',vmin = -1, vmax = 1)
plt.show()
```

# 가격에 영향을 적게 주는 컬럼 삭제

```python
df4 = df3.drop(['도담동', '보람동', '소담동', '소정면', '연기면', '연서면', '장군면', '전동면', '전의면'], axis=1)
```

# 최종 heatmap

```python
plt.figure(figsize=(16,14))
sns.heatmap(df4.corr(), annot=True, cmap = 'RdYlBu_r',vmin = -1, vmax = 1)
plt.show()
```
